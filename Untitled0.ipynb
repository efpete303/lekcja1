```python
data_loadtxt = np.loadtxt('example2.csv', delimiter=",", dtype=str)
print("Data loaded using txt:")
print(data_loadtxt[:5])
```

```python
data_genfromtxt = np.genfromtxt("example2.csv", delimiter=",",dtype=str, filling_values="N/A", encoding="utf-8")
print("Data loaded using genfromtxt:")
print(data_genfromtxt[:5])
```

```python
if data_genfromtxt.shape[1] >3:
  modified_data = np.delete(data_genfromtxt, 3, axis=1)
  print("Data after removing column 3:")
  print(modified_data[:5])
else:
  print("Column 3 does not exist in the dataset")
```
