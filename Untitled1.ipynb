```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
x = df[['Annual Income (k$)' , 'Spending Score (1-100)']]
```

```python
x
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
inertia = []
K_range = range(1,11)
for k in K_range:
  model = KMeans(n_clusters=k)
  model.fit(x_scaled)
  inertia.append(model.inertia_)
```

```python
inertia
```

```python
for k, inert in zip(K_range, inertia):
  print(f"k = {k}, inertia = {inert}")
```

```python
plt.figure(figsize=(10,5))
plt.plot(K_range, inertia, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('inertia')
plt.title('Elbow method for optimal k')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
clusters = kmeans.fit_predict(x_scaled)
df['Cluster'] = clusters
```

```python
plt.figure(figsize=(8, 6))
import seaborn as sns
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='Set2')
plt.title('Customer Segments based on annual Income nd Spending score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```
